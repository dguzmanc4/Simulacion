```python
import random
import matplotlib.pyplot as plt

def roll_dice():
    return random.randint(1,6)

def dice_game(a):
    result = dict()
    for i in range(a):
        sum = roll_dice() + roll_dice()
        if sum in result:
            result[sum] = result[sum] + 1
        else:
            result[sum] = 1
    return result  

def generate_histogram(data):
    plt.bar(list(data.keys()), data.values(), color='darkgreen')
    plt.title(f"Histograma correspondinte a {sum(data.values())} lanzamientos")
    plt.ylabel('# de repeticiones')
    plt.xlabel('Resultado del lanzamiento')
    plt.show()
```

```python
generate_histogram(dice_game(100))
```

```python
generate_histogram(dice_game(1000))
```

```python
generate_histogram(dice_game(10000))
```
