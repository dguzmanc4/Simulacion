```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('Conformación de los Grupos Económicos y su Comportamiento Tributario (4).xlsx', header=[4])
```

**Generar gráfico de barras de solo las empresas que han mejorado y/o empeorado su ranking con respecto al año anterior.**

```python
ranking_2021 = df.iloc[:302, 2]

eje_x = ['Mejoran', 'Empeoran']
eje_y = [sum(map(lambda x : x > 0, ranking_2021)), sum(map(lambda x : x < 0, ranking_2021))]

plt.bar(eje_x, eje_y)
plt.ylabel('Número de empresas')
plt.title('Ranking 2021 con respecto al año 2020')
plt.show()
```

**Generar gráfico de las empresas que tienen panamá papers.**

```python
panama_papers = df.iloc[:302, 7]

tienen = sum(map(lambda x : x > 0, panama_papers))
no_tienen = len(panama_papers) - tienen

eje_x = ['SI', 'NO']
eje_y = [tienen, no_tienen]

plt.bar(eje_x, eje_y)
plt.ylabel('Número de empresas')
plt.ylabel('¿Tienen Panamá Paper?')
plt.title('Panamá Papers')
plt.show()
```

**Generar un pie de las empresas que más aportan a la economía del país.**

```python
impuestos_causado = df.iloc[:302, 20]
companies = df.iloc[:302, 4] 
top_companies = sorted(zip(impuestos_causado, companies), reverse=True)[:10]
total_impuestos = 1052497075

eje_y = [(lambda x: (x[0]*100)/total_impuestos)(x) for x in top_companies]
mylabels = [(lambda x: x[1].split('(')[0].strip())(x) for x in top_companies]

plt.figure(figsize=(15, 15))
plt.pie(eje_y, labels = mylabels)
plt.show() 
```

**Obtener las empresas que mayor patrimonio tienen (10).**

```python
patrimonio = df.iloc[:302, 35]
empresas = df.iloc[:302, 4] 

top_companies = sorted(zip(patrimonio, empresas), reverse=True)[:10]
eje_y = [(lambda x: x[0])(x) for x in top_companies]
eje_x = [(lambda x: x[1].split('(')[0].strip())(x) for x in top_companies]

plt.bar(eje_x, eje_y)
plt.xlabel('Empresa')
plt.ylabel('Patrimonio')
plt.title('Top empresas con mayor patrimonio')
plt.xticks(eje_x, eje_x, rotation=90)
plt.show()
```
